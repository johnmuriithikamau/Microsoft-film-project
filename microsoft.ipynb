# Data Preparation:

To prepare the raw data for analysis, the following steps were taken:

### Step 1: Data Collection
The raw data was collected from the selected sources, including `bom.movie_gross.csv.gz`, `tmdb.movies.csv.gz`, `imdb.title.ratings.csv.gz`, `bom.movie_gross.csv.gz`, `imdb.title.basics.csv.gz` and `tn.movie_budgets.csv.gz`.


```python
import csv
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Reading csv Files
df_budgets = pd.read_csv('tn.movie_budgets.csv.gz')
df_tmdb = pd.read_csv('tmdb.movies.csv.gz')
df_gross = pd.read_csv('bom.movie_gross.csv.gz')
df_basics = pd.read_csv('imdb.title.basics.csv.gz')
df_crew = pd.read_csv('imdb.title.crew.csv.gz')
df_ratings = pd.read_csv('imdb.title.ratings.csv.gz')
```


### Step 2: Data Integration
The collected data from different sources needed to be integrated into a single dataset for analysis. This involved merging the relevant features from each source based on common identifiers such as movie titles or unique IDs.




```python
# Join the DataFrames on the tconst column.
df_joined1 = pd.merge(df_basics, df_ratings, left_on='tconst', right_on='tconst')
# Join the resulting DataFrame with the df_crew DataFrame.
df_joined2 = pd.merge(df_joined1, df_crew, left_on='tconst', right_on='tconst')

df_joined2.head()
```

```python
# Join the DataFrames on the primary_title and title column.
df_combined = pd.merge(df_joined2, df_gross, left_on='primary_title', right_on='title')

df_combined.head()
```

## Step 3: Data Cleaning
#### Handling Missing Values:
 Missing values were identified in the dataset and appropriate methods were used to handle them. For numerical features, missing values were imputed with mean, median, or zero values, depending on the context. For categorical features, missing values were replaced with the most frequent category.






Checking missing values in the entire DataFrame

 I  clean the data by removing any rows with missing values. For example, the following code would remove any rows in the df_joined1 DataFrame with missing values in the `averagerating` column:

```python
# Remove any rows with missing values in the averagerating column.
df_joined2 = df_joined2.dropna(axis=0, how='any', subset=['averagerating'])
df_joined2
```

### Prepare the data for analysis.
By converting the data types of the columns to the appropriate types.

convert the `averagerating` column to a float:

```python
# Convert the averagerating column to a float.
df_joined2['averagerating'] = df_joined2['averagerating'].astype(float)

df_joined2
```

```python
# Remove the dollar signs from the `production_budget` and `worldwide_gross` columns.
df_budgets['production_budget'] = df_budgets['production_budget'].str.replace('$', '').str.replace(',', ' ').str.replace(' ', '')
df_budgets['worldwide_gross'] = df_budgets['worldwide_gross'].str.replace('$', '').str.replace(',', ' ').str.replace(' ', '')

df_budgets
```

```python
# Convert the `production_budget` and `worldwide_gross` columns to numbers.
df_budgets['production_budget'] = df_budgets['production_budget'].astype(float)
df_budgets['worldwide_gross'] = df_budgets['worldwide_gross'].astype(float)
```

What is the distribution of ratings for movies?



```python
# Plot the distribution of ratings.
plt.hist(df_joined1['averagerating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movie Ratings')
plt.show()
```

What are the most popular movies?

```python

# Get the number of votes for each movie.
df_joined2['numvotes'] = df_joined2['numvotes'].astype(int)
# Sort the DataFrame by the number of votes.
df_joined2 = df_joined2.sort_values('numvotes', ascending=False)
# Get the top 10 movies.
top_10_movies = df_joined2.head(10)

top_10_movies
```

What are the most profitable movies?

```python
# Calculate the profit for each movie.
df_budgets['profit'] = df_budgets['worldwide_gross'] - df_budgets['production_budget']
# Sort the DataFrame by the profit.
df_budgets = df_budgets.sort_values('profit', ascending=False)
# Get the top 10 movies.
top_10_movies = df_budgets.head(10)

top_10_movies
```

# Data visualization

Are profit increasing on budget? 

```python
# Calculate the average budget profit.
profit_budget = df_budgets.groupby('production_budget')['profit'].mean()


# Plot the average budget profit.
plt.figure(figsize=(10, 6))
profit_budget .plot(kind='line')
plt.title('Average Profit Trends')
plt.xlabel('production budget')
plt.ylabel('profit')
plt.show()
```



Are ratings for movies increasing or decreasing over time?

```python

# Calculate the average rating for each year.
ratings_by_year = df_joined2.groupby('start_year')['averagerating'].mean()


# Plot the average rating over time.
plt.figure(figsize=(10, 6))
ratings_by_year.plot(kind='line')
plt.title('Average Ratings for Movies Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```



### Bar chart of the top 10 most profitable movies
This bar chart shows the top 10 most profitable movies, ranked by their worldwide gross. The chart is easy to read and understand, and it provides a clear picture of which movies have been the most successful at the box office.

```python

# Sort the DataFrame by profit in descending order
df_sorted = df_budgets.sort_values('profit', ascending=False)
# Select the top 10 movies
top_10_movies = df_sorted.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_movies['movie'], top_10_movies['profit'] / 1e9)  # Dividing by 1e9 for billions
plt.xlabel('Movie')
plt.ylabel('Profit (in billions)')
plt.title('Top 10 Most Profitable Movies')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

Pie chart of the top 5 genres by box office gross.

```python
# Group the DataFrame by genre and calculate the sum of box office gross for each genre
genre_gross = df_combined.groupby('genres')['domestic_gross'].sum().nlargest(5)
# Get the top 5 genres and their corresponding box office gross
top_genres = genre_gross.index
gross_values = genre_gross.values

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(gross_values, labels=top_genres, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Genres by Box Office Gross')
plt.axis('equal')
plt.show()
```
